### Preprccessing

```python
{column: len(data[column].unique()) for column in data.columns}
```

```python
def preprocess_inputs(df):
    df = df.copy()
    
    df = df.drop(['id'], axis=1)
    
    y = df['IsFraud'].copy()
    X = df.drop('IsFraud', axis = 1).copy()
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, random_state = 123)
    
    scaler = StandardScaler()
    scaler.fit(X_train)
    
    X_train = pd.DataFrame(scaler.transform(X_train), columns = X_train.columns)
    X_test = pd.DataFrame(scaler.transform(X_test), columns = X_test.columns)
    
    return X_train, X_test, y_train, y_test
```

```python
X_train, X_test, y_train, y_test = preprocess_inputs(data)
```

### Handling Class Imbalance

```python
train_df = pd.concat([X_train, y_train.reset_index(drop=True)], axis=1)
```

```python
oversampled_data = train_df.query("IsFraud == 0").sample(206, replace=False, random_state = 123)
```

```python
fraud_data = train_df.query("IsFraud == 1")
```

```python
train_df = pd.concat([fraud_data, oversampled_data], axis = 0).sample(frac=1.0, random_state = 123).reset_index(drop=True)
```

```python
y_train = train_df['IsFraud'].copy()
X_train = train_df.drop(['IsFraud'], axis=1).copy()
```

```python
X_train.shape
```

### PyCarret

```python
from pycaret.classification import *
```

```python
_ = setup(data=pd.concat([X_train, y_train], axis=1), target='IsFraud')
```

```python
model = compare_models()
```

### Modelling

```python
model.fit(X_train, y_train)
```

```python
tuned_model = tune_model(model)
```

### Results

```python
from sklearn.metrics import accuracy_score, roc_auc_score

y_pred1 = model.predict(X_test)
y_pred2 = tuned_model.predict(X_test)

# Calculate accuracy
accuracy1 = accuracy_score(y_test, y_pred1)
accuracy2 = accuracy_score(y_test, y_pred2)
print("Test Accuracy 1: {:.3f}%".format(accuracy1 * 100))
print("Test Accuracy 2: {:.3f}%".format(accuracy2 * 100))

# Calculate AUC score
auc_score1 = roc_auc_score(y_test, y_pred1)
auc_score2 = roc_auc_score(y_test, y_pred2)
print("     Test AUC 1: {:.3f}".format(auc_score1))
print("     Test AUC 2: {:.3f}".format(auc_score2))
```
