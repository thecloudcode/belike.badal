```python
train_cols = ['X_Minimum', 'X_Maximum', 'Y_Minimum', 'Y_Maximum',
       'Pixels_Areas', 'X_Perimeter', 'Y_Perimeter', 'Sum_of_Luminosity',
       'Minimum_of_Luminosity', 'Maximum_of_Luminosity', 'Length_of_Conveyer',
       'TypeOfSteel_A300', 'TypeOfSteel_A400', 'Steel_Plate_Thickness',
       'Edges_Index', 'Empty_Index', 'Square_Index', 'Outside_X_Index',
       'Edges_X_Index', 'Edges_Y_Index', 'Outside_Global_Index', 'LogOfAreas',
       'Log_X_Index', 'Log_Y_Index', 'Orientation_Index', 'Luminosity_Index',
       'SigmoidOfAreas']
target_cols = ['Pastry', 'Z_Scratch', 'K_Scatch', 'Stains',
       'Dirtiness', 'Bumps', 'Other_Faults']
```

```python
train = df[train_cols]
train.shape
```

```python
test.shape
```

```python
target = df[target_cols]
```

```python
skewed_cols_train = []

for i in train.columns:
    skew = scipy.stats.skew(train[i])
    if abs(skew)>=0.5:
        skewed_cols_train.append(i)
        print(i, skew, train[i].min())
```

```python
for i in skewed_cols_train:
    train[i] = np.log1p(train[i])
```

```python
skewed_cols_test = []

for i in test.columns:
    skew = scipy.stats.skew(test[i])
    if abs(skew)>=0.5:
        skewed_cols_test.append(i)
        print(i,skew)
```

```python
for i in skewed_cols_test:
    test[i] = np.log1p(test[i])
```

```python
for i in skewed_cols_test:
    if i not in skewed_cols_train:
        print(i)
```

```python
# Pastry

from sklearn.ensemble import GradientBoostingRegressor
pastry_reg = GradientBoostingRegressor(random_state=1024)
pastry_reg.fit(train, pd.DataFrame(target['Pastry']))
```

```python
# Z_Scratch

from lightgbm import LGBMRegressor
Z_Scratch_reg = GradientBoostingRegressor(random_state=5497)
Z_Scratch_reg.fit(train, pd.DataFrame(target['Z_Scratch']))
```

```python
# K_Scatch

K_Scatch = GradientBoostingRegressor(random_state=6503)
K_Scatch.fit(train, pd.DataFrame(target['K_Scatch']))
```

```python
# Stains

from sklearn.ensemble import ExtraTreesRegressor

Stains = ExtraTreesRegressor(n_jobs=-1, random_state=6639)
Stains.fit(train, pd.DataFrame(target['Stains']))
```

```python
# Dirtiness

Dirt = GradientBoostingRegressor(random_state=3316)
Dirt.fit(train, pd.DataFrame(target['Dirtiness']))
```

```python
# Bumps

Bumps = GradientBoostingRegressor(random_state=5048)
Bumps.fit(train, pd.DataFrame(target['Bumps']))
```

```python
# Other_Faults


of = GradientBoostingRegressor(random_state=4859)
of.fit(train, pd.DataFrame(target['Other_Faults']))
```

```python
s['Pastry'] = pastry_reg.predict(test)
s['Z_Scratch'] = Z_Scratch_reg.predict(test)
s['K_Scatch'] = K_Scatch.predict(test)
s['Stains'] = Stains.predict(test)
s['Dirtiness'] = Dirt.predict(test)
s['Bumps'] = Bumps.predict(test)
s['Other_Faults'] = of.predict(test)
```
