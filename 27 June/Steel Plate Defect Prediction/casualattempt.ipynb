```python
train_cols = ['X_Minimum', 'X_Maximum', 'Y_Minimum', 'Y_Maximum',
       'Pixels_Areas', 'X_Perimeter', 'Y_Perimeter', 'Sum_of_Luminosity',
       'Minimum_of_Luminosity', 'Maximum_of_Luminosity', 'Length_of_Conveyer',
       'TypeOfSteel_A300', 'TypeOfSteel_A400', 'Steel_Plate_Thickness',
       'Edges_Index', 'Empty_Index', 'Square_Index', 'Outside_X_Index',
       'Edges_X_Index', 'Edges_Y_Index', 'Outside_Global_Index', 'LogOfAreas',
       'Log_X_Index', 'Log_Y_Index', 'Orientation_Index', 'Luminosity_Index',
       'SigmoidOfAreas']
target_cols = ['Pastry', 'Z_Scratch', 'K_Scatch', 'Stains',
       'Dirtiness', 'Bumps', 'Other_Faults']
```

```python
train = df[train_cols]
train.shape
```

```python
test.shape
```

```python
target = df[target_cols]
```

```python
skewed_cols_train = []

for i in train.columns:
    skew = scipy.stats.skew(train[i])
    if abs(skew)>=0.5:
        skewed_cols_train.append(i)
        print(i, skew, train[i].min())
```

```python
for i in skewed_cols_train:
    train[i] = np.log1p(train[i])
```

```python
skewed_cols_test = []

for i in test.columns:
    skew = scipy.stats.skew(test[i])
    if abs(skew)>=0.5:
        skewed_cols_test.append(i)
        print(i,skew)
```

```python
for i in skewed_cols_test:
    test[i] = np.log1p(test[i])
```

```python
for i in skewed_cols_test:
    if i not in skewed_cols_train:
        print(i)
```

```python
# Pastry

from sklearn.ensemble import ExtraTreesClassifier
pastry_reg = ExtraTreesClassifier(bootstrap=False, ccp_alpha=0.0, class_weight=None,
                     criterion='gini', max_depth=None, max_features='sqrt',
                     max_leaf_nodes=None, max_samples=None,
                     min_impurity_decrease=0.0, min_samples_leaf=1,
                     min_samples_split=2, min_weight_fraction_leaf=0.0,
                     monotonic_cst=None, n_estimators=100, n_jobs=-1,
                     oob_score=False, random_state=8584, verbose=0,
                     warm_start=False)
pastry_reg.fit(train, pd.DataFrame(target['Pastry']))
```

```python
# Z_Scratch

from sklearn.ensemble import GradientBoostingClassifier
Z_Scratch_reg = GradientBoostingClassifier(ccp_alpha=0.0, criterion='friedman_mse', init=None,
                           learning_rate=0.1, loss='log_loss', max_depth=3,
                           max_features=None, max_leaf_nodes=None,
                           min_impurity_decrease=0.0, min_samples_leaf=1,
                           min_samples_split=2, min_weight_fraction_leaf=0.0,
                           n_estimators=100, n_iter_no_change=None,
                           random_state=5495, subsample=1.0, tol=0.0001,
                           validation_fraction=0.1, verbose=0,
                           warm_start=False)
Z_Scratch_reg.fit(train, pd.DataFrame(target['Z_Scratch']))
```

```python
# # K_Scatch

from sklearn.ensemble import RandomForestClassifier
K_Scatch = RandomForestClassifier(bootstrap=True, ccp_alpha=0.0, class_weight=None,
                       criterion='gini', max_depth=None, max_features='sqrt',
                       max_leaf_nodes=None, max_samples=None,
                       min_impurity_decrease=0.0, min_samples_leaf=1,
                       min_samples_split=2, min_weight_fraction_leaf=0.0,
                       monotonic_cst=None, n_estimators=100, n_jobs=-1,
                       oob_score=False, random_state=4580, verbose=0,
                       warm_start=False)
K_Scatch.fit(train, pd.DataFrame(target['K_Scatch']))
```

```python
# # Stains

# from sklearn.ensemble import ExtraTreesRegressor

Stains = RandomForestClassifier(bootstrap=True, ccp_alpha=0.0, class_weight=None,
                       criterion='gini', max_depth=None, max_features='sqrt',
                       max_leaf_nodes=None, max_samples=None,
                       min_impurity_decrease=0.0, min_samples_leaf=1,
                       min_samples_split=2, min_weight_fraction_leaf=0.0,
                       monotonic_cst=None, n_estimators=100, n_jobs=-1,
                       oob_score=False, random_state=1323, verbose=0,
                       warm_start=False)
Stains.fit(train, pd.DataFrame(target['Stains']))
```

```python
# # Dirtiness

Dirt = RandomForestClassifier(bootstrap=True, ccp_alpha=0.0, class_weight=None,
                       criterion='gini', max_depth=None, max_features='sqrt',
                       max_leaf_nodes=None, max_samples=None,
                       min_impurity_decrease=0.0, min_samples_leaf=1,
                       min_samples_split=2, min_weight_fraction_leaf=0.0,
                       monotonic_cst=None, n_estimators=100, n_jobs=-1,
                       oob_score=False, random_state=6386, verbose=0,
                       warm_start=False)
Dirt.fit(train, pd.DataFrame(target['Dirtiness']))
```

```python
# Bumps

Bumps = GradientBoostingClassifier(ccp_alpha=0.0, criterion='friedman_mse', init=None,
                           learning_rate=0.1, loss='log_loss', max_depth=3,
                           max_features=None, max_leaf_nodes=None,
                           min_impurity_decrease=0.0, min_samples_leaf=1,
                           min_samples_split=2, min_weight_fraction_leaf=0.0,
                           n_estimators=100, n_iter_no_change=None,
                           random_state=197, subsample=1.0, tol=0.0001,
                           validation_fraction=0.1, verbose=0,
                           warm_start=False)
Bumps.fit(train, pd.DataFrame(target['Bumps']))
```
