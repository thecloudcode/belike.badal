### Pastry -- Optuna

```python
# Gradient Boosting Regressor

def objective(trial):
    # Define the search space for hyperparameters
    params = {
        'n_estimators': trial.suggest_int('n_estimators', 50, 1000),
        'learning_rate': trial.suggest_loguniform('learning_rate', 0.001, 0.1),
        'max_depth': trial.suggest_int('max_depth', 3, 10),
        'min_samples_split': trial.suggest_int('min_samples_split', 2, 20),
        'min_samples_leaf': trial.suggest_int('min_samples_leaf', 1, 20),
        'max_features': trial.suggest_uniform('max_features', 0.1, 1.0)
    }
    
    # Create and fit the model with the current hyperparameters
    model = GradientBoostingRegressor(**params, random_state=1024)
    model.fit(trainx, targetx['Pastry'])
    
    # Evaluate the model
    predictions = model.predict(trainy)
    mse = mean_squared_error(targety['Pastry'], predictions)
    return mse
```

```python
sampler = TPESampler(seed=1)
study = optuna.create_study(study_name="Pastry", direction="maximize", sampler=sampler)
```

```python
study.optimize(objective, n_trials=200)
```

```python
print('Best parameters:', study.best_params)
print('Best value:', study.best_value)
print('Best trial:', study.best_trial)
```

### Stains -- Optuna

```python
from sklearn.ensemble import ExtraTreesRegressor
```

```python
# ExtraTreesRegressor

def objective(trial):
    n_estimators = trial.suggest_int("n_estimators", 100, 1000, step=100)
    max_depth = trial.suggest_int("max_depth", 5, 20)
    min_samples_split = trial.suggest_int("min_samples_split", 2, 10)
    min_samples_leaf = trial.suggest_int("min_samples_leaf", 1, 5)

  # Create the ExtraTreesRegressor model with tuned hyperparameters
    model = ExtraTreesRegressor(n_estimators=n_estimators,
                              max_depth=max_depth,
                              min_samples_split=min_samples_split,
                              min_samples_leaf=min_samples_leaf,
                              n_jobs=-1,
                              random_state=6639)

  # Train the model
    model.fit(trainx, targetx['Stains'])

  # Make predictions on the training data
    predictions = model.predict(trainy)

  # Calculate the mean squared error (MSE)
    mse = mean_squared_error(targety['Stains'], predictions)

  # Return the negative MSE (to minimize during optimization)
    return mse
```

```python
sampler = TPESampler(seed=1)
study = optuna.create_study(study_name="Stains", direction="maximize", sampler=sampler)
```

```python
study.optimize(objective, n_trials=200)
```

```python
print('Best parameters:', study.best_params)
print('Best value:', study.best_value)
print('Best trial:', study.best_trial)
```
