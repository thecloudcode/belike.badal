```python
skewed_cols_train = []

for i in train.columns:
    skew = scipy.stats.skew(train[i])
    if abs(skew)>=0.5:
        skewed_cols_train.append(i)
        print(i, skew, train[i].min())
```

```python
for i in skewed_cols_train:
    train[i] = np.log1p(train[i])
```

```python
skewed_cols_test = []

for i in test.columns:
    skew = scipy.stats.skew(test[i])
    if abs(skew)>=0.5:
        skewed_cols_test.append(i)
        print(i,skew)
```

```python
for i in skewed_cols_test:
    test[i] = np.log1p(test[i])
```

```python
for i in skewed_cols_test:
    if i not in skewed_cols_train:
        print(i)
```

### Pastry

```python
_ = setup(data=pd.concat([train, pd.DataFrame(target['Pastry'])], axis=1), target = 'Pastry')
```

```python
compare_models()
```

```python
import optuna
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import train_test_split

# Define the objective function to optimize
def objective(trial):
    # Define hyperparameters to tune
    params = {
        'n_estimators': trial.suggest_int('n_estimators', 50, 1000),
        'learning_rate': trial.suggest_loguniform('learning_rate', 0.001, 0.1),
        'max_depth': trial.suggest_int('max_depth', 3, 10),
        'min_samples_split': trial.suggest_int('min_samples_split', 2, 20),
        'min_samples_leaf': trial.suggest_int('min_samples_leaf', 1, 20),
        'subsample': trial.suggest_uniform('subsample', 0.5, 1.0),
        'max_features': trial.suggest_uniform('max_features', 0.1, 1.0)
    }
    
    # Initialize Gradient Boosting Regressor with current hyperparameters
    model = GradientBoostingRegressor(**params)
    
    # Split data into train and validation sets
    X_train, X_val, y_train, y_val = train_test_split(train, target['Pastry'], test_size=0.2, random_state=42)
    
    # Train the model
    model.fit(X_train, y_train)
    
    # Predict on validation set
    y_pred = model.predict(X_val)
    
    # Calculate ROC AUC score
    score = roc_auc_score(y_val, y_pred)
    
    return score

# Create a study object and optimize the objective function
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=100)

# Get the best hyperparameters
best_params = study.best_params
print("Best Hyperparameters:", best_params)

# Train the final model using the best hyperparameters
best_model = GradientBoostingRegressor(**best_params)
best_model.fit(train, target['Pastry'])
```

```python
pastry_model = best_model
```
